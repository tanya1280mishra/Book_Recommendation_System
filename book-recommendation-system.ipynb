# Popularity Based Recommender System

```python
ratings_with_name=ratingsdf.merge(booksdf,on='ISBN')
```

```python
ratings_with_name
```

When you perform the operation ratings_with_name.groupby('Book-Title').count()['Book-Rating'], the result is a Series. In this case, num_ratings_df is a one-dimensional array-like object that holds the counts of 'Book-Rating' for each 'Book-Title'.

The rename() method for a Series does not accept the columns argument. The columns argument is specific to DataFrames, which have named columns. Since a Series does not have columns (it has an index and values), using columns will raise a TypeError.


```python
num_ratings_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating']


#2. Convert Series to DataFrame:
#If you want to work with a DataFrame and want to rename the resulting column, you can reset the index:
num_ratings_df = num_ratings_df.reset_index()  
num_ratings_df.columns = ['Book-Title', 'num_ratings']  # Renaming the columns as needed
# num_ratings_df.rename(columns={'Book-Ratings':'num_ratings'},inplace=True)
num_ratings_df
```

```python
# Drop rows where 'Book-Ratings' is equal to the problematic string  
ratings_with_name = ratings_with_name[ratings_with_name['Book-Rating'] != '0590567330059056733005905673300590567330']  

# Optionally, if you want to ensure you're working with numeric types after dropping:  
ratings_with_name['Book-Rating'] = pd.to_numeric(ratings_with_name['Book-Rating'], errors='coerce')  
```

```python
avg_ratings_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()  
avg_ratings_df.rename(columns={'Book-Rating': 'avg_ratings'}, inplace=True)  
avg_ratings_df
```

```python
popular_df=num_ratings_df.merge(avg_ratings_df,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings')
popular_df
```

```python
popular_df = popular_df.merge(booksdf,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
popular_df
```

# Collaborative Filtering Based Recommander System

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
pdhe_likhe_users=x[x].index
```

```python
pdhe_likhe_users
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(pdhe_likhe_users)]
filtered_rating
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
print(famous_books)
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings.sort_values('Book-Rating',ascending=False)
```

# The values in the cells are the ratings that users assigned to the books. 
For instance:
If a user has rated "1984" with a score of 9.0, that means User-ID 254 rated the book "1984" as 9.
A value of 0.0 indicates that the user has not rated that particular book.

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True)
pt
```

 # Understanding Cosine Similarity
Cosine Similarity: This metric measures the cosine of the angle between two non-zero vectors. It is often used to determine how similar two items are, regardless of their magnitude. A cosine similarity of 1 indicates that the two items are identical in terms of their feature vectors, while a cosine similarity of 0 indicates that they are orthogonal (completely dissimilar).

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores=cosine_similarity(pt)
similarity_scores.shape
```

```python
def recommend(book_name):
    #index fetch
    index=np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data=[]
    for i in similar_items:
        item=[]
        temp_df=booksdf[booksdf['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    return data    
```

```python
recommend('1984')
```
